```python
# taking care of the parenthesis in the french targets list
for i, row in targets.iterrows():
    phrase = row['target']
    phrase = phrase.replace('"','')
    phrase = phrase.replace("''",'')
    phrase = str('« ') + phrase + str(' »')
    
    targets.at[i,'target'] = phrase
```

```python
#matching these conditions against the ones in amandine's 
def dic_creation(og_cond, target_cond, original_df = og_df, new_df = targets):
    
    #sentences
    #note: since every possible sentence occurs 2 (for each speaker, we can only collect it once for the dictionnary
    list1_eng = list(original_df[og_cond & (original_df['$speech_condition'] == 'french')]['$item'])
    list1_fr  = list(new_df[target_cond & (targets['$speech_condition'] == 'french')]['target'])

    #create dictionnary to replace items
    phrases = {}
    for i in list1_eng:
        for j in list1_fr:
            phrases[i] = j
            list1_fr.remove(j)
            break
    return phrases  #merging dics
```

```python
dic1 = dic_creation(cond1, target_cond1)
dic2 = dic_creation(cond2, target_cond2)
dic3 = dic_creation(cond3, target_cond3)
dic4 = dic_creation(cond4, target_cond4)

item_translations = dic1|dic2|dic3|dic4 #merging dics
#og_df.iloc[22]
len(item_translations) == 48*4 #checking that the number of item translations matches the number of items (4 cond for each 48 sentence)
```

```python
#replace them in the df
og_df = og_df.replace(item_translations)
```

```python
fillers = fillers.replace('. ', '.') #remove all spaces after point in sentences

og_df.keys()

filler_mask = (og_df["COLUMN: $sentence_type"] != 'target')

# Create a dictionary mapping English sentences to French sentences
print(len(fillers['english sentence']) == len(fillers['french sentence'])) #making sure there are the same number of stimuli

translation_item = dict(zip(fillers['english sentence'], fillers['french sentence']))
translation_para = dict(zip(fillers['english paraphrase'], fillers['french paraphrase']))

# Update the $item column in og_df based on the translation dictionary
og_df.loc[filler_mask,'$item'] = str('« ') + og_df['$item'].map(translation_item) + str(' »')
og_df.loc[filler_mask,'$paraphrase_validity'] = og_df['$paraphrase_validity'].map(translation_para)
```
